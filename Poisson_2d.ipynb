# PDE BIDIMENSIONAL (ec. POISON aplicada a la Transferencia de Calor)

Considere el problema de hallar la función $u(x,y)$ que satisfaga la ecuación diferencial parcial (Ec. de poison):
$$
$$
$$
-\frac{\partial ^2u}{\partial x^2}-\frac{\partial ^2u}{\partial y^2}=f(x,y)
$$

Se considera que el dominio es una placa cuadrada Unitaria y las condiciones aplicadas sobre el contorno se indican en la figura.


Utilizando Diferencias centrales,
$$
\frac{\partial^2 u}{\partial x^2}=\frac{T_{i+1,j}-2.T_{i,j}+T_{i-1,j}}{\Delta h_{x}^2} + o(h_{x}^2)
$$
$$
$$
$$
\frac{\partial^2 u}{\partial y^2}=\frac{T_{i,j+1}-2.T_{i,j}+T_{i,j-1}}{\Delta h_{y}^2}+ o(h_{y}^2)
$$

Despreciando los términos de orden superior y desarrollando un esquema implícito, y suponiendo además que $\Delta h_{x} = \Delta h_{y} = h$

Despreciando los términos de orden superior $o(h_{x,y}^2)$ y Reemplazando se obtiene,
$$
$$
$$
-\frac{T_{i+1,j}-2.T_{i,j}+T_{i-1,j}}{\Delta h_{x}^2} -\frac{T_{i,j+1}-2.T_{i,j}+T_{i,j-1}}{\Delta h_{y}^2} = f(x,y)
$$
$$
$$
Reordenando,
$$
$$
$$
T_{i+1,j} + T_{i-1,j} + T_{i,j+1} + T_{i,j-1} - 4.T_{i,j} = -f(x,y)*h^2
$$

# 1. Matriz de coeficientes del problema (Resolución solver JUPYTER)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import mpl_toolkits.mplot3d

#------------------------------------------
#------- Constantes y Cond. Inic. ---------
#------------------------------------------ 
L=1.0            # Largo de la placa
Np=30            # Cantidad de elementos por lado
dx=L/Np          # Delta posición   dx = dy

Nodos=(Np+1)**2            # Cantidad de Nodos
m=np.zeros((Nodos,Nodos))  # Inicializo Matriz de coeficientes
R=np.zeros(Nodos)          # Inicializo Vector de resultados
#xx=np.linspace(0.0,1.0,30)
xx = np.linspace(0.0, L, Np+1) 
yy = np.linspace(0.0, L, Np+1) 
X, Y = np.meshgrid(xx, yy) # Grilla x-y

T1,T2,T3,T4=300.0,300.0,1000.0,300 # Temp. Impuestas Bordes (Ver figura)

#------------------------------------------
#---------- Ensamble MATRIZ  --------------
#------------------------------------------
for j in range(1,Np+2):      # Recorre las FILAS
    for k in range(1,Np+2):  # Recorre las COLUMNAS
        p=(j-1)*(Np+1) + k -1
        f=X[j-1,k-1]**2 + Y[j-1,k-1]**2
            
        #-------------------------------
        #--------- PRIMER FILA ---------
        #-------------------------------            
        if j==1:
            R[p]=T4
            m[p,p]=1.0
            
        #-------------------------------
        #-------- ULTIMA FILA ----------
        #-------------------------------
        elif j==Np+1:
            R[p]=T2
            m[p,p]=1.0
                          
        #-------------------------------
        #----- ANTE ULTIMA FILA --------
        #-------------------------------
        elif j==Np:
            if k==1:
            #---primer columna---
                R[p]=T1
                m[p,p]=1.0
            elif k==Np+1:
            #---ultima columna---
                R[p]=T3
                m[p,p]=1.0
            elif k==2:
            #---segunda columna--- 
                m[p,p]=-4.0
                m[p,p+1]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]= -f*dx**2-T1 -T2 
            elif k==Np:
            #---Ante ultima columna--- 
                m[p,p]=-4.0
                m[p,p-1]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]= -f*dx**2 -T3 -T2 
            else: 
            #---Columnas intermedias--- 
                m[p,p]=-4.0
                m[p,p-1]=1.0
                m[p,p+1]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]= -f*dx**2 -T2
            
            #-------------------------------
            #------ FILAS INTERMEDIAS ------
            #-------------------------------
        else:               
            if k==1:
            #---primer columna---
                R[p]=T1
                m[p,p]=1.0
            elif k==Np+1:
            #---ultima columna---
                R[p]=T3
                m[p,p]=1.0
            elif k==2:
            #---segunda columna--- 
                m[p,p]=-4.0
                m[p,p+1]=1.0
                m[p,p+Np+1]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]=-f*dx**2 -T1 
            elif k==Np:
            #---Ante ultima columna--- 
                m[p,p]=-4.0
                m[p,p-1]=1.0
                m[p,p+Np+1]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]= -f*dx**2 -T3 
            else: 
            #---Columnas intermedias--- 
                m[p,p]=-4.0
                m[p,p-1]=1.0
                m[p,p+1]=1.0
                m[p,p+(Np+1)]=1.0
                m[p,p-(Np+1)]=1.0
                R[p]= -f*dx**2

print('la matriz de coeficientes es:',m)

Resul=np.linalg.inv(m).dot(R)  
T=Resul

Z=T.reshape(Np+1,Np+1)   # Rediseño el vector TEMPERATURA NODAL

cs = plt.contourf(X, Y, Z) 
cbar = plt.colorbar(cs) 
plt.title('POISSON 2D - Solver JUPYTER') 
plt.show() 
```

# 2. Resolución por método de JACOBI

```python
def Jacobi(Nelementos,MaxIteraciones,Tolerancia):
    tol=Tolerancia
    Maxiter=MaxIteraciones
    #------------------------------------------
    #------- Constantes y Cond. Inic. ---------
    #------------------------------------------ 
    L=1.0            # Largo de la placa
    Np=Nelementos    # Cantidad de elementos por lado
    dx=L/Np          # Delta posición   dx = dy
    Nodos=(Np+1)**2  # Cantidad de Nodos
    T1,T2,T3=300.0,300.0,1000.0 # Temp. Impuestas Bordes (Ver figura)

    k=0
    error=1.0
    err=np.zeros(Nodos)
    Eh=np.zeros(Maxiter+1)
   
    xx = np.linspace(0.0, L, Np+1) 
    yy = np.linspace(0.0, L, Np+1) 
    X1, Y1 = np.meshgrid(xx, yy) # Grilla x-y

    #---------------------------------------------
    #----------- Inicializo variables ------------
    #---------------------------------------------
    du=np.zeros(Nodos) # Inicializo du
    u=np.zeros(Nodos)  # Inicializo U

    while error>=tol and k<=Maxiter:
        # sumatoria U (k+1)= U(k) + du(k) 
        for m in range(1,Nodos+1):
            u[m-1]=u[m-1]+du[m-1]; 
        
        k=k+1
        for i in range(1,Np+2):      # Recorre las FILAS
            for j in range(1,Np+2):  # Recorre las COLUMNAS
                p=(i-1)*(Np+1) + j -1
                f=X1[i-1,j-1]**2 + Y1[i-1,j-1]**2
                
                #-------------------------------
                #--------- PRIMER FILA ---------
                #-------------------------------            
                if i==1:
                    du[p]=0.0
                    u[p]=T4
            
                #-------------------------------
                #-------- ULTIMA FILA ----------
                #-------------------------------
                elif i==Np+1:
                        du[p]=0.0
                        u[p]=T2
                          
                #-------------------------------
                #----- FILAs INTERMEDIAS -------
                #-------------------------------
                else:
                    if j==1:
                    #---primer columna---
                        du[p]=0.0
                        u[p]=T1
                    elif j==Np+1:
                    #---ultima columna---
                        du[p]=0.0
                        u[p]=T3
                    else: 
                    #---Columna intermedia--- 
                        du[p]=0.25*(u[p-1] + u[p+1] + u[p +Np+1] + u[p -(Np+1)]- 4*u[p] + f*dx**2)

        err=np.absolute(du)
        error=np.max(err)   
        Eh[k-1]=error
    return u, du, Eh, k, error, X1, Y1
```

##  Método Jacobi  ( h = 1 /10 )

```python
N=10          # Cantidad de elementos por Lado
Maxiter=6500  # Máximas iteraciones
Tol=1e-8      # Tolerancia

[u, du, Eh, k, error, X1, Y1]=Jacobi(N,Maxiter,Tol) # LLamo función


print('La cantidad de elementos por Lado es: ',N)
print('El error máximo es: ',error)
print('La cantidad de iteraciones Necesarias fueron:',k)
ee1=Eh  # Guardo el historial del error 
k1=k

Z1=u.reshape(N+1,N+1)   # Rediseño el vector TEMPERATURA NODAL

cs = plt.contourf(X1, Y1, Z1) 
cbar = plt.colorbar(cs) 
plt.title('POISSON 2D - Método Jacobi  ( h=1/10 )') 
plt.show() 
```

# 3. Resolución por método de Gauss Seidel 

El Método de Gauss Seidel para la resolución del sistema de ecuaciones se representa mediante el siguiente algoritmo

$$
x_{i}^{k+1} = \frac{1}{a_{i i}}.(b_{i} - \sum_{j=1}^{i-1} a_{i j}.x_{i}^{k+1} - \sum_{j=i+1}^{n} a_{i j}.x_{i}^{k})
$$

```python
def Gauss_Seidel(Nelementos,MaxIteraciones,Tolerancia):
    tol=Tolerancia
    Maxiter=MaxIteraciones
    #------------------------------------------
    #------- Constantes y Cond. Inic. ---------
    #------------------------------------------ 
    L=1.0            # Largo de la placa
    Np=Nelementos    # Cantidad de elementos por lado
    dx=L/Np          # Delta posición   dx = dy
    Nodos=(Np+1)**2  # Cantidad de Nodos
    T1,T2,T3,T4=300.0,300.0,1000.0,300.0 # Temp. Impuestas Bordes (Ver figura)

    k=0
    error=1.0
    err=np.zeros(Nodos)
    Eh=np.zeros(Maxiter+1)
   
    xx = np.linspace(0.0, L, Np+1) 
    yy = np.linspace(0.0, L, Np+1) 
    X1, Y1 = np.meshgrid(xx, yy) # Grilla x-y

    #---------------------------------------------
    #----------- Inicializo variables ------------
    #---------------------------------------------
    du=np.zeros(Nodos) # Inicializo dU
    u=np.zeros(Nodos)  # Inicializo U
    u1=np.zeros(Nodos) # Inicializo U en k+1

    while error>=tol and k<=Maxiter:
        u=u1
        k=k+1
        for i in range(1,Np+2):      # Recorre las FILAS
            for j in range(1,Np+2):  # Recorre las COLUMNAS
                p=(i-1)*(Np+1) + j -1
                f=X1[i-1,j-1]**2 + Y1[i-1,j-1]**2
                #f=0.0
                #-------------------------------
                #--------- PRIMER FILA ---------
                #-------------------------------            
                if i==1:
                    u[p]=T4
                    u1[p]=T4
                        
                #-------------------------------
                #-------- ULTIMA FILA ----------
                #-------------------------------
                elif i==Np+1:
                        u[p]=T2
                        u1[p]=T2
                          
                #-------------------------------
                #----- FILAs INTERMEDIAS -------
                #-------------------------------
                else:
                    if j==1:
                    #---primer columna---
                        u[p]=T1
                        u1[p]=T1
                    elif j==Np+1:
                    #---ultima columna---
                        u[p]=T3
                        u1[p]=T3
                    elif j==2:
                    #---Segunda columna---
                        du[p]=0.25*(u[p-1] + u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p -(Np+1)])
                        u1[p]=u[p] + du[p]
                    else: 
                    #---Columna intermedia--- 
                        du[p]=0.25*(u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p-1] + u1[p -(Np+1)] )
                        u1[p]=u[p] + du[p]

        err=np.absolute(du)
        error=np.max(err)   
        Eh[k-1]=error
    return u1, Eh, k, error, X1, Y1
```

## Gauss Seidel ( h = 1 / 10)

```python
N=10          # Cantidad de elementos por Lado
Maxiter=6500  # Máximas iteraciones
Tol=1e-8      # Tolerancia

[ug, Ehg, kg, errorg, X1g, Y1g]=Gauss_Seidel(N,Maxiter,Tol) # LLamo función


print('La cantidad de elementos por Lado es: ',N)
print('El error máximo es: ',errorg)
print('La cantidad de iteraciones Necesarias fueron:',kg)
ee4=Ehg  # Guardo el historial del error 
k4=kg

Z1g=ug.reshape(N+1,N+1)   # Rediseño el vector TEMPERATURA NODAL

cs = plt.contourf(X1g, Y1g, Z1g) 
cbar = plt.colorbar(cs) 
plt.title('POISSON 2D - Gauss Seidel  ( h=1/10 )') 
plt.show() 
```

# 4. Resolución por el Método SOR (sobre relajaciones sucesivas)

El Método de SOR para la resolución del sistema de ecuaciones se representa mediante el siguiente algoritmo, donde $w$ es un parámetro de relajación del método,

$$
x_{i}^{k+1} = \frac{w}{a_{i i}}.({b_{i} - \sum_{j=1}^{i-1} a_{i j}.x_{i}^{k+1} - \sum_{j=i+1}^{n} a_{i j}.x_{i}^{k}}) + (1-w).x_{i}^{k}
$$

```python
def Sor(Nelementos,MaxIteraciones,Tolerancia,w):
    tol=Tolerancia
    Maxiter=MaxIteraciones
    #------------------------------------------
    #------- Constantes y Cond. Inic. ---------
    #------------------------------------------ 
    L=1.0            # Largo de la placa
    Np=Nelementos    # Cantidad de elementos por lado
    dx=L/Np          # Delta posición   dx = dy
    Nodos=(Np+1)**2  # Cantidad de Nodos
    T1,T2,T3,T4=300.0,300.0,1000.0,300.0 # Temp. Impuestas Bordes (Ver figura)

    k=0
    error=1.0
    err=np.zeros(Nodos)
    Eh=np.zeros(Maxiter+1)
   
    xx = np.linspace(0.0, L, Np+1) 
    yy = np.linspace(0.0, L, Np+1) 
    X1, Y1 = np.meshgrid(xx, yy) # Grilla x-y

    #---------------------------------------------
    #----------- Inicializo variables ------------
    #---------------------------------------------
    du=np.zeros(Nodos) # Inicializo dU
    u=np.zeros(Nodos)  # Inicializo U
    u1=np.zeros(Nodos) # Inicializo U en k+1

    while error>=tol and k<=Maxiter:
        u=u1
        k=k+1
        for i in range(1,Np+2):      # Recorre las FILAS
            for j in range(1,Np+2):  # Recorre las COLUMNAS
                p=(i-1)*(Np+1) + j -1
                f=X1[i-1,j-1]**2 + Y1[i-1,j-1]**2
                #f=0.0
                #-------------------------------
                #--------- PRIMER FILA ---------
                #-------------------------------            
                if i==1:
                    u[p]=T4
                    u1[p]=T4
                        
                #-------------------------------
                #-------- ULTIMA FILA ----------
                #-------------------------------
                elif i==Np+1:
                        u[p]=T2
                        u1[p]=T2
                          
                #-------------------------------
                #----- FILAs INTERMEDIAS -------
                #-------------------------------
                else:
                    if j==1:
                    #---primer columna---
                        u[p]=T1
                        u1[p]=T1
                    elif j==Np+1:
                    #---ultima columna---
                        u[p]=T3
                        u1[p]=T3
                    elif j==2:
                    #---Segunda columna---
                        du[p]=w*0.25*(u[p-1] + u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p -(Np+1)])
                        u1[p]=(1.0-w)*du[p] + du[p]*w +u[p]
                    else: 
                    #---Columna intermedia--- 
                        du[p]=w*0.25*(u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p-1] + u1[p -(Np+1)] )
                        u1[p]=(1.0-w)*du[p] + du[p]*w +u[p]

        err=np.absolute(du)
        error=np.max(err)   
        Eh[k-1]=error
    return u1, Eh, k, error, X1, Y1
```

## SOR ( óptimo valor de  " w " )

```python
N=[10,20,40]  # Cantidad de elementos por Lado
Maxiter=6500  # Máximas iteraciones
Tol=1e-8      # Tolerancia

W=[0.8,1.0,1.4,1.6,1.7,1.8,1.9,2.0]
k7=np.zeros((8,3))
for j in range(1,4):
    for i in range(1,9):
        [uS, EhS, kS, errorS, X3S, Y3S]=Sor(N[j-1],Maxiter,Tol,W[i-1]) # LLamo función
        k7[i-1,j-1]=kS

h=DataFrame({'w':W,' h = 1 / 10 ':k7[:,0],' h = 1 / 20 ':k7[:,1],' h = 1 / 40 ':k7[:,2]})
h

```

# 5. Resolución por el Método SSOR 


El método **SSOR** (Symmetric successive over relaxation) consiste en dos iteraciones **SOR**, una en una dirección y la otra en la dirección inversa,


```python
def Ssor(Nelementos,MaxIteraciones,Tolerancia,w):
    tol=Tolerancia
    Maxiter=MaxIteraciones
    #------------------------------------------
    #------- Constantes y Cond. Inic. ---------
    #------------------------------------------ 
    L=1.0            # Largo de la placa
    Np=Nelementos    # Cantidad de elementos por lado
    dx=L/Np          # Delta posición   dx = dy
    Nodos=(Np+1)**2  # Cantidad de Nodos
    T1,T2,T3,T4=300.0,300.0,1000.0,300.0 # Temp. Impuestas Bordes (Ver figura)

    k=0
    error=1.0
    err=np.zeros(Nodos)
    Eh=np.zeros(Maxiter+1)
   
    xx = np.linspace(0.0, L, Np+1) 
    yy = np.linspace(0.0, L, Np+1) 
    X1, Y1 = np.meshgrid(xx, yy) # Grilla x-y

    #---------------------------------------------
    #----------- Inicializo variables ------------
    #---------------------------------------------
    du=np.zeros(Nodos)  # Inicializo dU  en (k+0.5)
    du1=np.zeros(Nodos) # Inicializo dU  en (k + 1)
    u=np.zeros(Nodos)   # Inicializo U
    u1=np.zeros(Nodos)  # Inicializo U en (k+0.5)
    u2=np.zeros(Nodos)  # Inicializo U en (k + 1)

    while error>=tol and k<=Maxiter:
        u=u2
        k=k+1
       
       #----------------------------------------
       #------- Ciclo hacia la Derecha --------- 
       #----------------------------------------
        for i in range(1,Np+2):      # Recorre las FILAS
            for j in range(1,Np+2):  # Recorre las COLUMNAS
                p=(i-1)*(Np+1) + j -1
                f=X1[i-1,j-1]**2 + Y1[i-1,j-1]**2

                #-------------------------------
                #--------- PRIMER FILA ---------
                #-------------------------------            
                if i==1:
                    u[p]=T4
                    u1[p]=T4
                        
                #-------------------------------
                #-------- ULTIMA FILA ----------
                #-------------------------------
                elif i==Np+1:
                        u[p]=T2
                        u1[p]=T2
                          
                #-------------------------------
                #----- FILAs INTERMEDIAS -------
                #-------------------------------
                else:
                    if j==1:
                    #---primer columna---
                        u[p]=T1
                        u1[p]=T1
                    elif j==Np+1:
                    #---ultima columna---
                        u[p]=T3
                        u1[p]=T3
                    elif j==2:
                    #---Segunda columna---
                        du[p]=w*0.25*(u[p-1] + u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p -(Np+1)])
                        u1[p]=(1.0-w)*du[p] + du[p]*w +u[p]
                    else: 
                    #---Columna intermedia--- 
                        du[p]=w*0.25*(u[p+1] + u[p +Np+1] - 4*u[p] + f*dx**2 + u1[p-1] + u1[p -(Np+1)] )
                        u1[p]=(1.0-w)*du[p] + du[p]*w +u[p]
        
        
        p=0
        f=0.0
       #----------------------------------------
       #------- Ciclo hacia la izquierda ------- 
       #----------------------------------------                 
        for i in range(Np+1,1,-1):      # Recorre las FILAS
            for j in range(Np+1,1,-1):  # Recorre las COLUMNAS
                p=(i-1)*(Np+1) + j -1
                f=X1[i-1,j-1]**2 + Y1[i-1,j-1]**2

                #-------------------------------
                #--------- PRIMER FILA ---------
                #-------------------------------            
                if i==1:
                    du1[p]=T4
                    u2[p]=T4
                        
                #-------------------------------
                #-------- ULTIMA FILA ----------
                #-------------------------------
                elif i==Np+1:
                        u2[p]=T2
                          
                #-------------------------------
                #----- FILAs INTERMEDIAS -------
                #-------------------------------
                else:
                    if j==1:
                    #---primer columna---
                        u2[p]=T1
                    elif j==Np+1:
                    #---ultima columna---
                        u2[p]=T3
                    elif j==2:
                    #---Segunda columna---
                        du1[p]=w*0.25*(u1[p-1] + u2[p+1] + u2[p +Np+1] - 4*u1[p] + f*dx**2 + u1[p -(Np+1)])
                        u2[p]=(1.0-w)*du1[p] + du1[p]*w +u1[p]
                    else: 
                    #---Columna intermedia--- 
                        du1[p]=w*0.25*(u2[p+1] + u2[p +Np+1] - 4*u1[p] + f*dx**2 + u1[p-1] + u1[p -(Np+1)] )
                        u2[p]=(1.0-w)*du1[p] + du1[p]*w +u1[p]                  
                                                

        err=np.absolute(du1)
        error=np.max(err)   
        Eh[k-1]=error
    return u2, Eh, k, error, X1, Y1
```

```python
N=[10,20,40]  # Cantidad de elementos por Lado
Maxiter=6500  # Máximas iteraciones
Tol=1e-8      # Tolerancia

W=[0.8,1.0,1.4,1.6,1.7,1.8,1.9,2.0]
k8=np.zeros((8,3))
for j in range(1,4):
    for i in range(1,9):
        [uS, EhS, kS, errorS, X3S, Y3S]=Ssor(N[j-1],Maxiter,Tol,W[i-1]) # LLamo función
        k8[i-1,j-1]=kS

h=DataFrame({'w':W,' h = 1 / 10 ':k8[:,0],' h = 1 / 20 ':k8[:,1],' h = 1 / 40 ':k8[:,2]})
h

```
